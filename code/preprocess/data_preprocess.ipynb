# Data Preprocessing for UK Housing Price Analysis

This notebook performs initial data preprocessing on the raw UK housing dataset.

**Tasks performed:**
- Add unique `House_Id`
- Calculate `House_Age` (based on year 2025)
- Calculate `Price_Per_Sqft`
- Reorder columns for better readability
- Save the processed dataset to `data/processed/`

**Output file:** `house_price_reg_processed.csv`

```python
import pandas as pd
import numpy as np
from pathlib import Path

# Set display options
pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.2f}'.format)

print("Libraries imported successfully.")
```

```python
# Define project paths (relative to this notebook)
raw_data_path = Path('../../data/raw/house_price_regression_dataset.csv')
processed_data_path = Path('../../data/processed/house_price_reg_processed.csv')

# Create directories if they don't exist
processed_data_path.parent.mkdir(parents=True, exist_ok=True)

print(f"Raw data path: {raw_data_path}")
print(f"Processed data will be saved to: {processed_data_path}")
```

```python
# Load the raw dataset
df = pd.read_csv(raw_data_path)

print(f"Raw dataset loaded successfully!")
print(f"Shape: {df.shape[0]:,} rows × {df.shape[1]} columns")
print("\nFirst 5 rows:")
display(df.head())
```

```python
# Make a copy to avoid modifying original data
df_processed = df.copy()

# 1. Add House_Id as unique identifier
df_processed['House_Id'] = range(1, len(df_processed) + 1)

# 2. Calculate House_Age (based on 2025)
df_processed['House_Age'] = (2025 - df_processed['Year_Built']).astype(int)

# 3. Calculate Price_Per_Sqft
df_processed['Price_Per_Sqft'] = (df_processed['House_Price'] / df_processed['Square_Footage']).round(2)

print("Feature engineering completed:")
print(f"   - Added House_Id")
print(f"   - Added House_Age (reference year: 2025)")
print(f"   - Added Price_Per_Sqft")
```

```python
# Define desired column order
desired_column_order = [
    'House_Id', 
    'Year_Built', 
    'House_Age',
    'House_Price', 
    'Price_Per_Sqft',
    'Square_Footage', 
    'Num_Bedrooms', 
    'Num_Bathrooms',
    'Lot_Size', 
    'Garage_Size', 
    'Neighborhood_Quality'
]

# Reorder columns
df_processed = df_processed[desired_column_order]

print("Columns reordered successfully.")
print("\nFinal column order:")
print(df_processed.columns.tolist())
```

```python
print("=== Final Dataset Overview ===")
print(f"Total observations: {len(df_processed):,}")
print(f"Total features: {df_processed.shape[1]}")

print("\nData types:")
print(df_processed.dtypes)

print("\nMissing values:")
print(df_processed.isnull().sum())

print("\nSummary statistics:")
display(df_processed.describe().round(2))

print("\nFirst 5 rows of processed data:")
display(df_processed.head())
```

```python
# Save to processed folder
df_processed.to_csv(processed_data_path, index=False)

print(f"Processed dataset successfully saved to:")
print(f"   {processed_data_path}")
print(f"   Shape: {df_processed.shape[0]:,} rows × {df_processed.shape[1]} columns")
```
